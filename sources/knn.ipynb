```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv ('https://raw.githubusercontent.com/plotly/datasets/master/iris-data.csv')
```

```python
X = dataset.iloc[:,0:4]
y = dataset.iloc[:,4]
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
dataset['class']= label_encoder.fit_transform(dataset['class']) 
dataset['class'].unique()
label_encoder_y = LabelEncoder()
y = label_encoder_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train,y_train)
```

```python
hasil_prediksi = classifier.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy= accuracy_score(y_test, hasil_prediksi)*100
print('Akurasi: ' + str(round(accuracy, 2)) + '%')
```
